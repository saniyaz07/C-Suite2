```python
new_df=df[['Song-Name','Mood','Language','tags']]
new_df 
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df['Mood'] = new_df['Mood'].str.lower()
new_df['Language'] = new_df['Language'].str.lower()

new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True ,key=(lambda x:x[1]))
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
    m_i=new_df[new_df['title']==music].index[0]
    d=similarity[m_i]
    m_l=sorted(list(enumerate(d)),reverse=True,key=lambda x:x[1])[1:6]
    for i in m_l:
        print(new_df.iloc[i[0]].title)
recommend('Bom Diggy Diggy')        
```
